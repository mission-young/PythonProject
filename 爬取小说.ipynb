# 爬取八一中文网小说

```python
import requests
from bs4 import BeautifulSoup
link='http://zwdu.com'
nextlink='/book/24390/7357689.html'
f=open('./最强升级系统.txt','w',encoding='utf-8')
print("######################\n######################\n最强升级系统下载中···")
bookid=1
while not nextlink == '/book/24390/':
    print('第%d章下载中' % bookid)
    links=link+nextlink
    res=requests.get(links)
    res.encoding='gbk'
    soup=BeautifulSoup(res.text,'html.parser')
    nextlink=soup.select('.bottem1')[0].select('a')[2].get('href')
    contents=soup.select('#content')[0].text
    contents=contents.replace('㈧』㈠中┡ 』文网ＷｗんＷ．８⒈Ｚｗ．ＣＯＭ','')
    contents=contents.replace('㈧㈠Δ』中文网Ｗｗ％Ｗ．ん８⒈Ｚｗ．ＣＯＭ','')
    contents=contents.replace('㈧㈠中文网Ｗｗ『Ｗ．％８⒈Ｚｗ．ＣＯＭ','')
    contents=contents.replace('火热免费新书，喜欢的兄弟请下载，谢谢！','')
    contents=contents.replace('——','- - - - - - - - \n\n\n\n')
    contents=contents.replace('    ','\n')
    f.write(contents)
    bookid=bookid+1
    f.flush()
f.close()
```

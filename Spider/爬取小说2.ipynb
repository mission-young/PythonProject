# 爬取手机小说下载网小说

```python
import requests
from bs4 import BeautifulSoup
link='http://www.sjtxt.la/book/36717/'
nextlink='2807767.html'
f=open('./镇魂.txt','w',encoding='utf-8')
print("######################\n######################\n镇魂下载中···")
bookid=1
while not nextlink == '/book/36717/':
    print('第%d章下载中' % bookid)
    links=link+nextlink
    res=requests.get(links)
    res.encoding='utf-8'
    soup=BeautifulSoup(res.text,'html.parser')
    nextlink=soup.select('.txt_lian')[1].select('a')[3].get('href')
    contents=soup.select('#content1')[0].text
#     contents=contents.replace('㈧』㈠中┡ 』文网ＷｗんＷ．８⒈Ｚｗ．ＣＯＭ','')
#     contents=contents.replace('㈧㈠Δ』中文网Ｗｗ％Ｗ．ん８⒈Ｚｗ．ＣＯＭ','')
#     contents=contents.replace('㈧㈠中文网Ｗｗ『Ｗ．％８⒈Ｚｗ．ＣＯＭ','')
#     contents=contents.replace('火热免费新书，喜欢的兄弟请下载，谢谢！','')
#     contents=contents.replace('——','- - - - - - - - \n\n\n\n')
    contents=contents.replace('    ','\n')
    f.write(contents)
    bookid=bookid+1
    f.flush()
f.close()
print("mission complete!")
```
